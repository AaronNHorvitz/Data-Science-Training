# Events and Sample Space

## Definitions

<p>
<dt>Event:</dt>
<dd>An <b>event</b> is a collection of outcomes of a procedure. (Ex. Flipping a coin, rolling the dice.)</dd>
</p>
<p>
<dt>Simple Event:</dt>
<dd>A <b>simple event</b> is a single outcome. (The number of ways in which you can accomplish the the outome or in some cases, not accomplish it.)</dd>
</p>
<p>
<dt>Sample Space:</dt>
<dd>The <b>sample space</b> All the simple events.(Is every possible outcome)</dd>    
</p>

# Probability 

## Definitions

<p>
<b>Probabilty</b> is the likelihood of an <b>event</b> occuring.
</p>

### Notation


<center>$\Large Prob:  P$<center>

<center>$\Large Event:  A$<center>

<center>$\Large P(A) = $ <font size=5>The Probability of Event "A" Occuring</font><center>
    

### 3 Types of Probability
- Observed Probability
- Classical Probability
- Subjective Probability (Talking to an "expert")

#### Observed Probability
Are you doing something? (This is what did happen.)

<p>
An <b>observed probability</b> is a probability that is estimated based on your observation.
</p>
<p>
</p> 
<center>$\large P(A) = \large \frac{\text{# of times "A" occured}}{\text{# of times a procedure was repeated}}$ <font size=3><center>
    


<b>Example: Batting Averages in Baseball</b>
- Batting average of 100 is 1 in 10 times the batter will hit the ball
- Batting average of 400 is 4 in 10 times the batter will hit the ball


#### Classical Probability
Are you just thinking about doing something? (This is what should happen)

<p>
<b>Classical probability</b> is based more on theory. It is the probability of a chance of an event occuring. (Each sample event must have an equal chance of occuring.)
</p>

<b>Example: A Weighted Die Won't Work</b>
- What happens if I give you a weighted die, and I ask you to tell me the odds of a six coming up? You don't know what the chances are of one side coming up because you don't know the weights. You are assuming there is an equal chance on all sides. So there has to be equal weights for this to work. 

<center>$\large P(A) = \large \frac{\text{# of times "A" COULD occur}}{\text{# of simple events (outcomes)}}$ <font size=3><center>

## Probabilities of 0 and 1

### Zero Probability
Events that are impossible have zero probability, but the converse is not necessarily true. Clearly an empty set has zero probability. But if you are calculating your probability using an observations, then a zero probability event does not mean an impossible event.

<b>Example</b> Flipping a coin:
- If you have flipped a coin ten(10) times and it comes up heads each time. It's not certain that your next flip will be heads as well.

### One Probability

A probability of 1 means it is a certain event. 

## Example with code - Using coin flipping to find the odds

It never quite gets to a perfect even split

```python
import random 
from tqdm import tqdm

# n is the number of iterations
n = 1000000

zero_count = 0
one_count = 0

for count in tqdm(range(n), desc = 'Loading...'):
    
    value = random.randint(0, 1)
    if value == 0:
        zero_count += 1
    else:
        one_count += 1
        
print(f'{zero_count} heads and {one_count} tails')
```

The larger the number of iterations, the closer it gets

```python
# n is the number of iterations
cycles = 1000

zero_count = 0
one_count = 0
differences = []  

for n in tqdm(range(1, cycles), desc = 'Loading...'):
    
    for count in range(0, n):

        value = random.randint(0, 1)
        if value == 0:
            zero_count += 1
        else:
            one_count += 1

    percentage_difference = abs(((zero_count-one_count)/zero_count)*100)
    print(f'{n:<5} cycles | {zero_count} heads and {one_count} tails | {abs(zero_count-one_count)} difference ({round(percentage_difference,3)})%')
```

```python
from matplotlib import pyplot as plt

# n is the number of iterations
cycles = 400

zero_count = 0
one_count = 0
differences = []

for n in tqdm(range(1, cycles), desc = 'Loading...'):
    
    for count in range(0, n):

        value = random.randint(0, 1)
        if value == 0:
            zero_count += 1
        else:
            one_count += 1

        percentage_difference = abs(((zero_count-one_count)/zero_count)*100)
        differences.append(percentage_difference)

plt.plot(differences)
plt.ylabel('Percentage Difference')
plt.xlabel('Number of Cycles')
plt.axhline(y=0,color = 'red', linestyle = '--')
plt.show()
plt.close()
```

# Complementary Events

## Definitions

<p>
<dt>Mutually Exclusive:</dt>
<dd>You can't be in more than one group at the same time. <i>related in such a way that each thing makes the other thing impossible : not able to be true at the same time or to exist at the same time</i></dd>
</p>
<p>
<dt>Collectively Exhaustive:</dt>
<dd>A set of events (sample space) is jointly or collectively exhaustive if at least one of the events must occur. </dd>
</p>
<p>
<dt>Complementary Events:</dt>
<dd>Events which are <b>mutually exclusive</b>. The two events can't happen at the same time.</dd>    
</p>

<b>EXAMPLE:</b>  When rolling a six sided. Rolling a 2 and a 5 are mutually exclusive events

The compliment of event $\large A$ is denoted by $\large \bar{A}$ and is all the outcomes when event $\large A$ does not occur. 

<b>EXAMPLE</b>

 What is the complement of rolling a 5?

$\large \bar{A} = {1, 2, 3, 4 , 6}$

What is the probability of rolling a 5?

<center>$\large P(5) = \large \frac{\text{1}}{\text{6}}$<center>

What is the probability of not rolling a 5?

<center>$\large P(\bar5) = \large \frac{\text{5}}{\text{6}}$<center>
