### For Loop and Range
The for loop iterates through a series of numbers established by the <b>range</b> function. The range function has the following parameters.  The format is the following:
<b>range(start,stop,step)</b>

- <b>start</b>  -- OPTIONAL integer number specifying at which position to start. Default is 0
- <b>stop</b>   -- REQUIRED integer number specifiying at which position to end. 
- <b>step</b>   -- OPTIONAL number specifying the incrementation.  Default is 1. 

NOTE:  The only required variable is the stop variable.  However, the default for stop and step are set to 0 and 1 respectively.



```python
for number in range(1,10,.5):
    print(number)
```

### Exhuastive Enumeration - Guess And Check With a For Loop
This is an algorithmic technique that enumerates all possibilities until we get to the right answer or exhaust the space of possibilities. 
NOTE:  The <b>break</b> command allows us to exit the <b>for</b> loop before it's finished looping.

```python
#Find the square root of a perfect cube

number  = int(input('Enter a positive integer: '))
guess   = 0

start = 0
stop  = abs(number)
step  = 1

if number < 0:
    print(number, "is a negative number. I can't find the square root of a negative number.")
    

else: #Iterate through all the guesses. 

    for guess in range(start, stop, step):
    
        if guess**2 >= abs(number):
            break        

if guess**2 != abs(number):
    print(number, 'is not a perfect square.')
    
else:
    print('Square root of', number, 'is', guess)
```

## Caution About Using Float Numbers in For and While Loops!

Most of the time, number of type float provide a reasonably good approximation to real numbers.  But "most of the time" is not all of the time, and when they don't it can lead to surprising consequences.  

### Accumulation of Rounding Errors in Test for Equality

```python
#Consider the following code:

float_number = 0.0

for count in range (10):
    
    float_number = float_number + 0.1

if float_number == 1.0:
    
    print(float_number, '= 1.0')
else: 
    print(float_number, ' is not 1.0')
```

<p>To understand why this happens, we need to understand how floating point numbers are represented in the computer during computation.  Computers use electricity to work.  Inside a microchip, electricity is turned either <b>on</b> or <b>off</b>. This is represented by the symbols <b>1</b> or <b>0</b>. These are <b>binary numbers.</b> The computer uses binary numbers, not decimal numbers to do computation.  Therefore, to understand that, we need to understand <b>binary numbers.</b></p>
<p>A binary number unlike decimal numbers, are in base 2, not in base 10.  They are represented with either a 0 or a 1.  These digits are often called <b>bits</b>.  The right most digit is the 2<sup>0</sup> place the next digit towards the left is the 2<sup>1</sup> place, etc. </p>

-  0 0 0 0 0 0 <b>1</b>  -- has numerical value 2<sup>0</sup> = 1
-  0 0 0 0 0 <b>1</b> 0  -- has numerical value 2<sup>1</sup> = 2
-  0 0 0 0 <b>1</b> 0 0  -- has numerical value 2<sup>2</sup> = 4
-  0 0 0 <b>1</b> 0 0 0  -- has numerical value 2<sup>3</sup> = 8
-  0 0 <b>1</b> 0 0 0 0  -- has numerical value 2<sup>4</sup> = 16
-  0 <b>1</b> 0 0 0 0 0  -- has numerical value 2<sup>5</sup> = 32
-  <b>1</b> 0 0 0 0 0 0  -- has numerical value 2<sup>6</sup> = 128
  
<p>Therefore:</p><p></p>  <b>10011011</b>   has the value <b>1 x [128]</b> + 0 x [64] + 0 x [32] + <b>1 x [16]</b> + <b>1 x [8]</b> + 0 X [4] + <b>1 x [2]<b> + <b>1 x [1]</b> = 155

<p>Non integer numbers are represented using <b>floating point</b>.  We would represent the numbers as a pair of integers. These numbers are the <b>significant digits</b> of a number and an <b>exponent</b>.</p>

<p>Modern computes use binary, not decimal represenations.  We represent the significant digits and exponents in binary rather than 10 to the exponent.  For example, the number 0.625(5/8) would be representated as the pair (101,-11); because 5/8 is .101 in binary adn -11 is the binary represntation of -3, the pair (101, -11) stand for 5 x 2 <sup>-3</sup> = 5/8 = 0.625</p>

<p>In most Python implementations there are 53 bits of precision available for floating point numbers.  So the significant digits stored for the decimal number number 0.1 will be:</p>
<p><b>0.0001100110011001100110011001100110011001100110011</b></p>
<p>That is the decimal number:</p>
<p><b>0.1000000000000000055511151231257827021181583404541015625</b></p>
<p>That is more digits than most people find useful, so Python keeps the number of digits manageable by displaying a rounded value instead.</p>
<p><b>1/10 or 0.1</b></p>

### Scoping
Are the variables within the scope of the fuction, are they global, or are they outside the function?


```python
def f(x):
    y = 1
    x = x + y
    print('x =', x)
    return x
```

```python
x = 3
y = 2
z = f(x)

print('z =', z)
print('x =', x)
print('y =', y)
```

At the call of f, the formal parameter x is locally bound to the value of the actual parameter x.  It is important to note that though the actual and formal parameters have the same name, they are not the same variable.  Each function defines a new <b>name space</b> also called a <b>scope</b>. The formal parameter x and the <b>local parameter</b> y that are used in f exist only within the scope of the definition f.  The assignment satatement x = x + y within the function body binds the local name x to the object 4.  The assignemnts in f have no effect at all on the binding of the names x and y that exist outside of the scope of f.

## Recursion
<b>Recursion</b> is a way of programming or coding a problem, in which a function calls itself one or more times in its body. Usually, it is returning the return value of this function call. If a function definition fulfils the condition of recursion, we call this function a recursive function.

<p>In general, a recursive definition is made up of two parts.  There is at least one <b>base case</b> that directly specifies that specifies the result for a special case, and thereis at least one <b>recursive (inductive) case</b> that defines the answer in terms of the answer to the question on some other input, typically a simpler version of the same problem. </p>


### Example of recursion with the Fibonacci Sequence

```python
def fib(n):
    """Assumes n int > = 0
    Returns Fibonacci of n"""
    if n == 0 or n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)

def test_fib(n):
    for i in range(0,n+1,1):
        print('fib of',i,'=',fib(i))
    
```

```python
test_fib(20)
```

## Towers of Hanoi

The objective of the puzzle is to move the entire stack to last rod, obeying the following simple rules:

- Only one disk can be moved at a time.
- Each move consists of taking the upper disk from one of the stacks and placing it on top of another stack or on an empty rod.
- No larger disk may be placed on top of a smaller disk.
With 3 disks, the puzzle can be solved in 7 moves. The minimal number of moves required to solve a Tower of Hanoi puzzle is 2n  1, where n is the number of disks.
