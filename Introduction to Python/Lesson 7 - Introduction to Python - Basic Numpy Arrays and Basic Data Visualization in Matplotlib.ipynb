## Array Slicing

<pre>
<font face = "Helvetica", size = 3>
<font size="4"><b>Array Slicing:</b></font>  We can take a "slice" of the array, and take out of it a "subarray", using the slice notation. The slice notation is marked by a colon [<b>:</b>] character inside square brackets.  

It looks like this:   <b>x[start:stop:step]</b>

If any of these are left blank, the defaults are <b>start = 0, stop = size of the dimension, step = 1</b>.

<b>IMPORTANT!</b> Array slices return views rather than copies of the slices they make. Any changes made to the slice will be made to the original array.  If you would like to make changes to the slice without making changes to the original array, you need to make a copy of the array or slice of that array. 

```python
letters = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']
X = np.array(letters)
X
```

```python
X[::]    # iterate over the array in sequence
```

```python
X[::2]    # skip ever other letter
```

```python
X[3:8]    # select index 3 through 8
```

```python
X[:-5]    # select 0 to -5
```

### for loop iteration 

```python
# iterate from index 1 to the end of the array

for index in range(1,len(X)+1):
    print(index, X[:index])
```

### Modifying existing copies of an array

```python
letters = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']

original = np.array(letters)
sliced     = original[0:3]

print(original)
print(sliced)
```

```python
sliced[0] = 0

print(original)
print(sliced)
```

```python
# print the original array and compare
print("original:",original)
```

### Making copies of the array

```python
letters = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']

original = np.array(letters)
sliced = original[0:3].copy()

print(original)
print(sliced)       #Now it's a copy of the original.
```

```python
# modify the subset of the array by replacing the first value of the slice with the value 0

sliced[0] = 0
print(original)
print(sliced)
```

# Plotting the Results with Matplotlib

SOURCE: https://matplotlib.org/

Matplotlib is one of the one of the most important and most widely used visualization libraries in Python.  It contains many useful modules. The one we will be starting with today is <b>pyplot.<b> 

```python
import numpy as np

from matplotlib import pyplot as plt
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(0,10,.1)
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

### add some color

```python
plt.plot(x,y, color = 'r')
plt.show()
```

### overlay the plots

```python
plt.plot(x,y, color = 'dodgerblue')
plt.scatter(x,y, color = 'green', marker='+')
plt.show()
```

### change the figure size to fit the page

```python
plt.figure(figsize = (10,5))
plt.plot(x,y, color = 'dodgerblue')
plt.scatter(x,y, color = 'red')
plt.show()
```

### add a horizontal line at the zero mark

```python
plt.figure(figsize = (18,7))
plt.plot(x,y, color = 'dodgerblue')
plt.scatter(x,y, color = 'red')
plt.axhline(y=0, color = 'black', lw=2)
plt.axvline(x=5, color = 'yellow', lw=2)
plt.show()
```

### reduce the line thickness


```python
plt.figure(figsize = (18,7))
plt.plot(x,y, color = 'dodgerblue', linewidth = 1)
plt.scatter(x,y, color = 'red')
plt.axhline(y=0, color = 'black', linewidth = 1)
plt.show()
```

### make the spots more transparent

```python
plt.figure(figsize = (18,7))
plt.plot(x,y, color = 'dodgerblue', linewidth = 1)
plt.scatter(x,y, color = 'red', alpha = 0.5)
plt.axhline(y=0, color = 'black', linewidth = 1)
plt.show()
```

### add a cosine wave

```python
y_cos = np.cos(x)

plt.figure(figsize = (18,7))

plt.plot(x,y_cos, color = 'green', linewidth = 1)
plt.plot(x,y, color = 'dodgerblue', linewidth = 1)

plt.scatter(x,y, color = 'red', alpha = 0.3)
plt.axhline(y=0, color = 'black', linewidth = 1)
plt.show()
```

### remove the red spots and change the line style

```python
plt.figure(figsize = (18,7))

plt.plot(x,y_cos, color = 'green', linestyle = '--', linewidth = 2, label = 'cosine')
plt.plot(x,y, color = 'dodgerblue', linestyle = ':', linewidth = 2, label = 'sine')

plt.axhline(y=0, color = 'black', linewidth = 1)
plt.legend(loc='best')
plt.show()
```

### add a legend to the graph with label() and legend() and increase the line thickness

```python
plt.figure(figsize = (18,7))

plt.plot(x,y_cos, color = 'green', linestyle = '--', linewidth = 5, label = 'cosine')
plt.plot(x,y, color = 'dodgerblue', linewidth = 5, label = 'sine')

plt.axhline(y=0, color = 'black', linewidth = 1)
plt.legend()
plt.show()
```

### add a title to the graph

```python
plt.figure(figsize = (18,7))

plt.plot(x,y_cos, color = 'green', linestyle = '--', linewidth = 5, label = 'cosine')
plt.plot(x,y, color = 'dodgerblue', linewidth = 5, label = 'sine')

plt.axhline(y=0, color = 'black', linewidth = 1)
plt.title('Sine and Cosine Waves', fontsize=18)
plt.legend()
plt.show()
```

### add x and y labels to the plot and change the fontsize

```python
plt.figure(figsize = (18,7))

plt.plot(x,y_cos, color = 'green', linestyle = '--', linewidth = 5, label = 'cosine')
plt.plot(x,y, color = 'dodgerblue', linewidth = 5, label = 'sine')

plt.axhline(y=0, color = 'black', linewidth = 1)
plt.title('\nSine and Cosine Waves\n', fontsize = 20)

plt.xlabel('\nx', fontsize = 20)
plt.ylabel('y', rotation = 0, fontsize = 20)

plt.legend(fontsize = 14)
plt.show()
```

### Increase the tick size

```python
plt.figure(figsize = (18,7))

plt.plot(x,y_cos, color = 'green', linestyle = '--', linewidth = 5, label = 'cosine')
plt.plot(x,y, color = 'dodgerblue', linewidth = 5, label = 'sine')

plt.axhline(y=0, color = 'black', linewidth = 1)
plt.title('\nSine and Cosine Waves\n', fontsize = 20)

plt.xlabel('\nx', fontsize = 20)
plt.ylabel('y', rotation = 0, fontsize = 20)

plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.legend(fontsize = 14)
plt.show()
```
